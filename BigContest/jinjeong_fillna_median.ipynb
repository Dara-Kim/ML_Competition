```python
import pandas as pd
import os
import json
import pickle
import json
import joblib
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python
loan_df=pd.read_csv('./loan_result.csv')
log_df=pd.read_csv('./log_data.csv')
user_df=pd.read_csv('./user_spec.csv')
```

```python
os.listdir()
```

```python
loan_df = loan_df.loc[loan_df['is_applied'].dropna(axis=0).index]
```

```python
loan_df['loan_limit']=loan_df['loan_limit'].fillna(loan_df['loan_limit'].median())
```

```python
loan_df['loan_rate']=loan_df['loan_rate'].fillna(loan_df['loan_rate'].median())
```

```python
data=loan_df
data
```

```python
data.drop(['loanapply_insert_time'], axis=1, inplace=True)
```

```python
data.isnull().sum()
```

```python
data.isna().sum()
```

```python
result=dict()
```

```python
acc_list=[]
f1_list=[]
confusion_list=[]

for i in range(100):
    data_set=data.sample(100_000)
    x, y = data_set.loc[:,data_set.columns!='is_applied'], data_set['is_applied']
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
    model = RandomForestClassifier(random_state=42)
    model.fit(x_train, y_train)
    y_pred=model.predict(x_test)
    acc=model.score(x_test,y_test)
    f1=f1_score(y_test,y_pred)
    con=confusion_matrix(y_test,y_pred)
    acc_list.append(acc)
    f1_list.append(f1)
    confusion_list.append(con)
    with open(f'./mydata/data_set_fill_med_{i}.txt','w') as f:
        f.write('\n'.join(map(str,data_set.index)))
    saved_model = pickle.dumps(model)
    joblib.dump(model,f'./model/data_set_fill_med_{i}.pkl')
    print(i,'done')
```

```python
result=dict()
result['acc']=acc_list
result['f1']=f1_list
```

```python
for i, cm in enumerate(confusion_list):
    d=ConfusionMatrixDisplay(cm,display_labels=[0,1])
    fig,ax=plt.subplots(figsize=(2,2))
    d.plot(cmap='Greys',ax=ax)
    plt.savefig(f'./visual/model{i}cm_fill_med_.png')
```

```python
with open('./json/result1_fill_med_.json','w') as f:
    json.dump(result,f)
```
