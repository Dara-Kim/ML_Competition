# 값의 분포 살펴보기

## user_spec

```python
is_numeric_dtype(user_df['birth_year'])
```

```python
null_col_list = list(user_df.columns[user_df.isnull().sum()!=0])
for col in null_col_list:
    print(col)
    if is_numeric_dtype(user_df[col]):
        print(f'mean : {user_df[col].mean()}')
        print(f'median : {user_df[col].median()}')
        plt.hist(user_df[col])
        plt.show()
    else:
        CountStatus = pd.value_counts(user_df[col].values, sort=True)
        CountStatus.plot.bar()
        plt.show()
```

## loan_result

```python
null_col_list = list(loan_df.columns[loan_df.isnull().sum()!=0])
for col in null_col_list:
    print(col)
    if is_numeric_dtype(loan_df[col]):
        print(f'mean : {loan_df[col].mean()}')
        print(f'median : {loan_df[col].median()}')
        plt.hist(loan_df[col])
        plt.show()
    else:
        CountStatus = pd.value_counts(loan_df[col].values, sort=True)
        CountStatus.plot.bar()
        plt.show()
```
