# 요약 정리

* train input의 파일 개수는 58개, test input 파일 개수는 6개

* train input의 data를 살펴본 결과 null값이 몇몇 존재했음

```python
wrong_train_list=[]
for index, file_name in enumerate(os.listdir('./train_input/'), 1):
    train_input_df = pd.read_csv(f'./train_input/{file_name}')
    a=[]
    for col in train_input_df.columns:
        a+=list(train_input_df[train_input_df[col].isnull()].index)
    a=set(a)
    print(f' <{index}번째 결측값> \nshape : {train_input_df.shape}\n결측값 개수 : {sum(train_input_df.isnull().sum())}\n결측 값 행개수 : {len(a)}')
    if train_input_df.shape[0]*0.7 < len(a):
        print('\033[95m' +'결측값 70% 이상' + '\033[0m')
        wrong_train_list.append(index)
    print()
```
