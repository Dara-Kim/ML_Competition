```python
import pandas as pd
import os
from supervised.automl import AutoML
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor
```

```python
tmp_df= pd.read_csv('./data/train_input/CASE_01.csv')
tmp_df2= pd.read_csv('./data/train_target/CASE_01.csv')
tmp_df3= pd.read_csv('./data/train_input/CASE_02.csv')
tmp_df4= pd.read_csv('./data/train_target/CASE_02.csv')
```

input 데이터 전처리

```python
tmp_df.fillna(value=0,inplace=True)
tmp_df['시간'] = pd.to_datetime(tmp_df['시간'])
tmp_df['date'] = tmp_df['시간'].dt.date
tmp_df3.fillna(value=0,inplace=True)
tmp_df3['시간'] = pd.to_datetime(tmp_df3['시간'])
tmp_df3['date'] = tmp_df3['시간'].dt.date
```

target 데이터 전처리

```python
tmp_df2= pd.read_csv('./data/train_target/CASE_01.csv')
absolute_value=1
list_tmp_df2=[]
tmp_df2["rate+1"]= tmp_df2['rate']+1
for i in list(tmp_df2["rate+1"]):
    absolute_value*=i
    list_tmp_df2.append(absolute_value)
tmp_df2['rate_based_on_init']=list_tmp_df2
tmp_df2['tmp']=[1]+list_tmp_df2[:-1]
tmp_df2['value']=tmp_df2['rate_based_on_init']-tmp_df2['tmp']

tmp_df4= pd.read_csv('./data/train_target/CASE_02.csv')
absolute_value=1
list_tmp_df4=[]
tmp_df4["rate+1"]= tmp_df4['rate']+1
for i in list(tmp_df4["rate+1"]):
    absolute_value*=i
    list_tmp_df4.append(absolute_value)
tmp_df4['rate_based_on_init']=list_tmp_df4
tmp_df4['tmp']=[1]+list_tmp_df4[:-1]
tmp_df4['value']=tmp_df4['rate_based_on_init']-tmp_df4['tmp']
```

```python
tmp_df4
```

```python
my_model = AutoML(results_path ='./hojun_4_model1',)
my_model.fit(tmp_df.groupby('date').sum(), tmp_df2['value'])
```

```python
my_model.score(tmp_df3.groupby('date').sum(), tmp_df4['value'])
```

```python
r2 = r2_score(my_model.predict(tmp_df3.groupby('date').sum()), tmp_df4['value'])
print(r2)
```

```python
print(*zip(my_model.predict(tmp_df3.groupby('date').sum()), tmp_df4['value']), sep='\n')
```

```python
r2 = r2_score(clf1.predict(tmp_df3.groupby('date').mean()), tmp_df4['value'])
print(r2)
```

```python
print(*zip(clf1.predict(tmp_df3.groupby('date').mean()), tmp_df4['value']), sep='\n')
```
