데이터의 특성 파악을 해보았음

* input 데이터는 각 청경채에 대한 데이터로, 시간이 겹치는 것도 있지만, 아닌 것도 있음
* 각 데이터에 있는 결측값에는 규칙이 있다.
    * 결측값인 데이터의 개수는 2 혹은 37개이다
    * 2개의 결측값이 있는 케이스의 경우 모든 항의 데이터가 모두 2개의 결측값을 가진다.
        
        => 해당 2개의 결측값은 항상 외부온도추정관측치, 외부습도추정관측치이다
* 우선 target 데이터는 결측값이 없다.

그렇다면 어떻게 학습데이터를 만들어야할까..?

# 각 케이스의 시간은?

```python
import pandas as pd
import os
from tqdm import tqdm, notebook
```

```python
for filename in tqdm(os.listdir('./train_input/')):
    df = pd.read_csv(f'./train_input/{filename}')
    if '시간' not in df.columns:
        print(filename)
```

```python
time_list=[]
for filename in tqdm(os.listdir('./train_input/')):
    df = pd.read_csv(f'./train_input/{filename}')
    time_list+=list(df['시간'])
```

```python
def list_to_dict(list_a):
    dict_a = {}
    for elm in list_a:
        if elm in dict_a.keys():
            dict_a[elm] += 1
        else:
            dict_a[elm] = 1
    return dict_a
```

```python
time_dict = list_to_dict(time_list)
```

# 결측값은 몇개인가

```python
reuslt_dict=dict()
for filename in os.listdir('./train_input/'):
    df = pd.read_csv(f'./train_input/{filename}')
    reuslt_dict[filename]=list_to_dict([f'{v} nulls' for v in df.T.isnull().sum() if v>0])
    print(filename, reuslt_dict[filename], 'all_col_num :',df.shape[0])
```

```python
reuslt_dict=dict()
for filename in os.listdir('./test_input/'):
    df = pd.read_csv(f'./test_input/{filename}')
    reuslt_dict[filename]=list_to_dict([f'{v} nulls' for v in df.T.isnull().sum() if v>0])
    print(filename, reuslt_dict[filename], 'all_col_num :',df.shape[0])
```

# 2개의 결측값인 항목은 항상 같은 항목인가?

```python
for num in [37, 39, 43, 47, 49, 50, 54, 55]:
    df = pd.read_csv(f'./train_input/CASE_{num}.csv')
    sum_df=df.isnull().sum()
    print('df num :',num)
    for col in sum_df.index:
        if sum_df[col] == df.shape[0]:
            print(col)
```

```python
for num in [5]:
    df = pd.read_csv(f'./test_input/TEST_0{num}.csv')
    sum_df=df.isnull().sum()
    print('df num :',num)
    for col in sum_df.index:
        if sum_df[col] == df.shape[0]:
            print(col)
```

# target의 null 값

```python
for filename in os.listdir('./train_target/'):
    df = pd.read_csv(f'./train_target/{filename}')
    if df['rate'].isnull().sum() > 0 :
        print(filename, df['rate'].isnull().sum())
```
