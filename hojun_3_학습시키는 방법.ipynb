# 실험 3

```python
train_df = pd.DataFrame()
for filename in os.listdir('./data/train_input/')[:5]:
    df = pd.read_csv(f'./data/train_input/{filename}')
    df.fillna(value=0,inplace=True)
    df['시간'] = pd.to_datetime(df['시간'])
    df['date'] = df['시간'].dt.date
    df = df.groupby('date').sum()
    train_df = pd.concat([train_df,df])
```

```python
target_df = pd.DataFrame()
for filename in os.listdir('./data/train_target/')[:5]:
    df = pd.read_csv(f'./data/train_target/{filename}')
    target_df = pd.concat([target_df,df['rate']])
```

```python
my_model3 = AutoML()
my_model3.fit(train_df,target_df)
```

```python
def list_to_dict(list_a):
    dict_a = {}
    for elm in list_a:
        if elm in dict_a.keys():
            dict_a[elm] += 1
        else:
            dict_a[elm] = 1
    return dict_a
```
