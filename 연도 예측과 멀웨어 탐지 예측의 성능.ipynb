```python
df_dict= dict()
for i,v in enumerate(product( map(str, range(14,22)),'bm')):
    file=os.listdir('./csv')[i]
    key = ''.join(v)
    df_dict[key]= pd.read_csv(f'./csv/{file}', index_col=0)
    print(key, file)
```

```python
name_dict=dict()
for key in df_dict.keys():
    if key[2] == 'b':
        name_dict[key] = '20'+key[:2]+'년 Benign'
    else:
        name_dict[key] = '20'+key[:2]+'년 Malware'
```

```python
df_dict['14b'].iloc[1000:2000]
```

```python
ytrain  = [14]*8000 + [15]*8000 + [16]*8000 + [17]*8000 + [18]*8000 + [19]*8000 + [20]*8000 + [21]*8000
ytest  = [14]*2000 + [15]*2000 + [16]*2000 + [17]*2000 + [18]*2000 + [19]*2000 + [20]*2000 + [21]*2000
for i in range(5):
    Xtrain, Xtest=pd.DataFrame(), pd.DataFrame()
    for key in tqdm(df_dict.keys()):
        for step in range(5):
            if i == step:
                Xtest = pd.concat( [Xtest, df_dict[key].iloc[step*1000:(step+1)*1000]])
            else:
                Xtrain = pd.concat( [Xtrain, df_dict[key].iloc[step*1000:(step+1)*1000]])
    exec(f'model{i} = RandomForestClassifier(random_state=42).fit(Xtrain, ytrain)')
    print('model',i)
    print('accuracy', eval(f'model{i}').score(Xtrain, ytrain), eval(f'model{i}').score(Xtest, ytest))
#     print('f1score', f1_score(ytrain, eval(f'model{i}').predict(Xtrain)), f1_score(ytest, eval(f'model{i}').predict(Xtest)))
    print()
    time.sleep(2)
```

```python
result_df = pd.DataFrame()
for year, key in enumerate(df_dict.keys(), 28):
    print(year//2, key)
    for step in range(5):
        train_result=[]
        test_result=[]
        for i in range(5):
            if i==step:
                test_result.append(list(eval(f'model{i}').predict(df_dict[key].iloc[step*1000:(step+1)*1000])))
            else:
                train_result.append(list(eval(f'model{i}').predict(df_dict[key].iloc[step*1000:(step+1)*1000])))
        result_df = pd.concat([result_df, pd.DataFrame([[year//2]*1000]+train_result+test_result).T])
    
    
#     result_table.insert(0,[year//2]*5000)
#     result_df = pd.concat([result_df, pd.DataFrame(result_table).T])
result_df.columns=['label','train1','train2','train3','train4','test']
```

```python
ytrain  = [0]*4000 + [1]*4000
ytest  = [0]*1000 + [1]*1000
key_list = list(df_dict.keys())
for i in range(5):
    for year_i, year in enumerate(range(14,22)):
        Xtrain, Xtest=pd.DataFrame(), pd.DataFrame()
        key = key_list[year_i*2]
        for step in range(5):
            if i == step:
                Xtest = pd.concat( [Xtest, df_dict[key].iloc[step*1000:(step+1)*1000]])
            else:
                Xtrain = pd.concat( [Xtrain, df_dict[key].iloc[step*1000:(step+1)*1000]])
        key = key_list[year_i*2+1]
        for step in range(5):
            if i == step:
                Xtest = pd.concat( [Xtest, df_dict[key].iloc[step*1000:(step+1)*1000]])
            else:
                Xtrain = pd.concat( [Xtrain, df_dict[key].iloc[step*1000:(step+1)*1000]])
        exec(f'bmodel{i}_{year} = RandomForestClassifier(random_state=42).fit(Xtrain, ytrain)')
        print(f'bmodel{i}_{year}', key)
        print('accuracy', eval(f'bmodel{i}_{year}').score(Xtrain, ytrain), eval(f'bmodel{i}_{year}').score(Xtest, ytest))
        print('f1score', f1_score(ytrain, eval(f'bmodel{i}_{year}').predict(Xtrain)), f1_score(ytest, eval(f'bmodel{i}_{year}').predict(Xtest)))
        print()
```
